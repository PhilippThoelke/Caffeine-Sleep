```python
import os
import pickle
import numpy as np
from scipy import io
from mne import viz, stats
from matplotlib import colors, pyplot as plt
```

```python
CAF_DOSE = 200
SIGNIFICANT_P = 0.005
AGE_GROUP = -1 # -1: all, 0: up to age 30, 1: from age 30

PROJECT_PATH = '..\\data'
FEATURE_PATH = 'C:\\Users\\Philipp\\Documents\\Caffeine\\Features{dose}\\Combined'.format(dose=CAF_DOSE)

STAGES = ['AWA', 'AWSL', 'NREM', 'REM']
BANDS = ['delta', 'theta', 'alpha', 'sigma', 'beta', 'low gamma']
CHANNELS = ['Fp1', 'Fp2', 'F3', 'F4', 'F7', 'F8', 'C3', 'C4', 'P3 ', 'P4',
            'O1', 'O2 ', 'T3', 'T4', 'T5', 'T6', 'Fz', 'Cz', 'Pz', 'Oz']
```

```python
sensor_pos = io.loadmat(os.path.join(PROJECT_PATH, 'Coo_caf'))['Cor'].T
sensor_pos = np.array([sensor_pos[1], sensor_pos[0]]).T
```

```python
age_suffix = ''
if AGE_GROUP == 0:
    age_suffix = '_age_t30'
elif AGE_GROUP == 1:
    age_suffix = '_age_f30'

with open(os.path.join(FEATURE_PATH, f'data_avg{age_suffix}.pickle'), 'rb') as file:
    data = pickle.load(file)
with open(os.path.join(FEATURE_PATH, f'labels_avg{age_suffix}.pickle'), 'rb') as file:
    labels = pickle.load(file)
with open(os.path.join(FEATURE_PATH, f'groups_avg{age_suffix}.pickle'), 'rb') as file:
    groups = pickle.load(file)
```

```python
results = {}
for stage in STAGES:
    results[stage] = {}
    for feature, current in data[stage].items():
        results[stage][feature] = stats.permutation_t_test(current[labels[stage]==1] - current[labels[stage]==0],
                                                           n_permutations=10000,
                                                           tail=0,
                                                           n_jobs=-1)
```

```python
colormap = 'coolwarm'

for stage in STAGES:
    plt.figure(figsize=(18, 5))
    plt.suptitle(stage, y=1.1, fontsize=20)
    
    highest = np.max([np.max(np.abs(results[stage][feature][0].mean(axis=0))) for feature in results[stage].keys()])
    vmin = -highest
    vmax = highest
    
    subplot_index = 1
    axes = []
    for feature in results[stage].keys():
        curr_sig = results[stage][feature][1]
        curr_t = results[stage][feature][0]
        
        ax = plt.subplot(2, 5, subplot_index)
        axes.append(ax)
        plt.title(feature)
        mask = curr_sig < SIGNIFICANT_P
        viz.plot_topomap(curr_t, sensor_pos, mask=mask, cmap=colormap, vmin=vmin, vmax=vmax, contours=False, show=False)
        subplot_index += 1
        
    norm = colors.Normalize(vmin=vmin,vmax=vmax)
    sm = plt.cm.ScalarMappable(cmap=colormap, norm=norm)
    sm.set_array([])
    plt.colorbar(sm, ax=axes, shrink=0.95, aspect=15)
    plt.savefig(f'..\\results\\diff_avg{CAF_DOSE}\\diff_avg_{stage}{age_suffix}')
    plt.show()
```
